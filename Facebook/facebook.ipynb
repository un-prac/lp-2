```python
import pandas as pd

# Load the dataset
df = pd.read_csv('dataset_Facebook.csv', sep=';')
df.head()
```

### a. Create data subsets

```python
# Creating two subsets
subset1 = df[['Type', 'Category', 'Post Month']].loc[:10]
subset2 = df[['Type', 'Total Interactions']].loc[:10]

display(subset1.head())
display(subset2.head())
```

### b. Merge Data

```python
# Merging the two subsets on 'Type'
merged_df = pd.merge(subset1, subset2, on='Type')
merged_df
```

### c. Sort Data

```python
# Sorting the merged dataframe by 'Total Interactions'
sorted_df = merged_df.sort_values(by='Total Interactions', ascending=False)
sorted_df.head()
```

### d. Transposing Data

```python
# Transposing the sorted dataframe
transposed_df = sorted_df.transpose()
transposed_df
```

### e. Shape and reshape Data

```python
# Shaping (finding dimensions)
print('Shape of dataframe:', subset1.shape)

# Reshaping (using melt to change from wide to long format)
reshaped_df = pd.melt(subset1, id_vars=['Type'], value_vars=['Category', 'Post Month'])
reshaped_df.head()
```
